```python
# generate 50 independent random variables from N(0,1) distribution.
x = np.random.normal(size=50)
x
```

```python
# create an array y by adding an independent N(50,1) random variable to each variable to x.
y = x + np.random.normal(loc=50, scale=1, size=50)
y
```

```python
# correlation between x and y
np.corrcoef(x,y)
```

## 1.5 Indexing Data

```python
A = np.array(np.arange(16)).reshape((4,4))
A
```

```python
A[1,2], A[1][2]
```

```python
# Indexing Rows, Columns and Submatrices
A[[1,3],:]
```

```python
A[:, [1,2]]
```

```python
A[[1,3],[0,2]]
```

Above we are getting 1d-array. what is does is that it fetches A[1,0] and A[3,2].

For fetching submatrix.

```python
A[[1,3]][:,[0,2]]
```

```python
# using np.ix_()
idx = np.ix_([1,3],[0,2,3])
A[idx]
```

Boolean Indexing

```python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1,3]] = True
keep_rows
```

Above matrix is same as [0,1,0,1]


```python
np.all(keep_rows == np.array([0,1,0,1]))
```

```python
A[[0,1,0,1]]
```

```python
# or
A[np.array([0,1,0,1])]
```

```python
# but
A[keep_rows] # => retrieves only those rows indexing with true i.e row 2 and row 3
```
